```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/DATA SET/raw_data.csv")
df
df["income"]

#data transfoming
df2=df.copy()
df2["tax"]=df2["income"].apply(lambda x: "20%" if x>=50000 else "10%")
df2
```

```python
#use of map
gender_map={"Male": "M","Female":"F","Unknown":"U"}
df2["gender"]=df2["gender"].map(gender_map)
```

```python
df2
df2=df2.assign(new_income=df2["income"] *1.1)
```

```python
df2
```

```python
#renmaing
df2.columns=["Id","Name","Age","Country","Gender","Income","Tax","New_Income"]
df2
```

```python
#renaming particular column
df2=df2.rename(columns={"Income":"Salary"})
df2
```

```python
df2.sort_values("Salary")
sorted_value=df2.sort_values("Salary",ascending=False)
sorted_value
sorted_value.reset_index()

#ranking
sorted_value["Ranking"]=sorted_value["Salary"].rank(ascending=False)
sorted_value
```
