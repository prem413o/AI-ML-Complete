```python
import matplotlib.pyplot as plt
```

```python
#real data
year=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232]

plt.plot(year,oscar_revenue)
plt.title("oscar movie revenue collection in each year")
plt.xlabel("Years")
plt.ylabel("oscar revenue")
```

```python
#non_oscar movies
non_oscar_movies=["dj","dhol","daku","puspa","raj"]
non_oscar_year=[2008,2009,2010,2011,2012]
non_revenue=[378,2788,829,185,275]

plt.plot(year,oscar_revenue,label="oscar winner")
plt.title("oscar movies vs non_oscar movies")
plt.plot(non_oscar_year,non_revenue,label="non_oscar winner")
plt.xlabel("years")
plt.ylabel("Revenue")
plt.legend()
```
