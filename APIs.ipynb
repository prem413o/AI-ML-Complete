```python
import requests
```

```python
URL="https://stephen-king-api.onrender.com/api/books"
res=requests.get(URL)
```

```python
json_data=res.json()
```

```python
import pandas as pd

df=pd.json_normalize(json_data["data"])
```

```python
df=df[['id','Year','Title']]
```
