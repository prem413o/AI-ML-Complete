```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[5,6,7,8]

plt.plot(x,y)
plt.show()
```

```python
#oscar movie
year=[2008,2009,2010,2011,2012]
oscar_revenue=[137,1001,200,490,550]
plt.plot(year,oscar_revenue)
plt.title("oscar movie revenue collection")
plt.xlabel("Years")
plt.ylabel("oscar_revenue")
```

```python
#non_oscar movie
years=[2008,2009,2010,2011,2012]
nonoscar_revenue=[150,300,450,670,230]
plt.plot(year,oscar_revenue,"o--r",label="oscar collection")
plt.plot(years,nonoscar_revenue,label="non oscar collection")
plt.title("oscar movie vs nonoscar movie")
plt.xlabel("years")
plt.ylabel("Revenue")
plt.legend()
```

```python
years=[2008,2009,2010,2011,2012]
nonoscar_revenue=[150,300,450,670,230]
plt.plot(year,oscar_revenue,label="oscar collection")
plt.plot(years,nonoscar_revenue,label="non oscar collection")
plt.title("oscar movie vs nonoscar movie")
plt.xlabel("years")
plt.ylabel("Revenue")
plt.legend()
plt.style.available

plt.grid()
plt.show()
plt.savefig("graph.png")
```
