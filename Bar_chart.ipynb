```python
import matplotlib.pyplot as plt
```

```python
#oscar-movies
year=[2008,2009,2010,2011,2012]
revenue=[1005,170,427,133,232]
plt.bar(year,revenue,color="green")
plt.title("oscar revenue collection")
plt.xlabel("year")
plt.ylabel("Value in $")

for i in range(len(year)):
  plt.text(year[i],revenue[i],str(revenue[i]),ha="center")
```

```python
year=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232]

years=[2008,2009,2010,2011,2012]
non_oscar_revenue=[378,2788,829,185,275]

plt.bar(year,oscar_revenue)
plt.bar(years,non_oscar_revenue)

plt.title("oscar_revenue vs non-oscar-revenue")
plt.xlabel("years")
plt.ylabel("Revenue collection in $ ")
```
